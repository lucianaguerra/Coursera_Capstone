## a) Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

## b) Read the tables at the wikipedia url

```python
url = 'http://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
df_wiki1 = pd.read_html(url)
df_wiki1[0]
```

## c) Create a data frame with the table we are interested

```python
df_wiki2 = df_wiki1[0]
df_wiki2
```

## d) Replace the "Not assigned" boroughs to "NaN"

```python
df_wiki2['Borough'].replace('Not assigned',np.nan, inplace=True)
df_wiki2
```

## e) Drop the rows where the borough was 'Not assigned'

```python
df_wiki2.dropna(axis=0,inplace=True)
df_wiki2
```

## f) Verify if there is a repeated 'Postal Code'
